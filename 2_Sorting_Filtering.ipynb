## Filter

```python
df_English = df[df["SchoolMajor"] == "English"]
print(df_English.head())
```

```python
type(df_English) # pandas.core.frame.DataFrame
```

```python
df_abv_30 = df[df["Age"] > 30]
df_abv_30.head()
```

```python
df_abv_30["Age"].hist()
```

## Filtering by multiple conditions

```python
df_abv30_Cpt = df[(df["Age"] > 30) & (df["SchoolMajor"] == "Computer Science")]
df_abv30_Cpt.head()
```

```python
df_abv30_or_cmpt = df[(df["Age"] > 30) | (df["SchoolMajor"] == "ComputerScience")]
df_abv30_or_cmpt.head()
```

```python
df_idx_age = df.set_index("Age")
df_idx_age.head()
```

## Select both row and columns for speed using "at"

### value = df.at[2, "A"]

```python

df.at[2, "Age"]
```

```python
## Select data based on labels

### loc method is label-based, allowing you to select data by row and column labels
```

```python
# To select rows where the "Age" column is greater than 30

df.loc[df["Age"] > 30].head()
```

```python
## To select rows 0 to 3 and columns "Age" and "Gender"

df.loc[0:3, ["Age", "Gender"]]
```
