## Segmentation by mulitple features

- segment data by two catogorical features, e.g., number of children, IssoftwareDev

- use a pivot table

.pivot_table()
- specify rows, column, and values.


```python
df_gpby_nbchd = df.groupby("NumberOfChildren")
```

```python
df.pivot_table(
    index = "NumberOfChildren", # row
    columns = "IsSoftwareDev",
    values = "HoursSpentLearningToCode",
    aggfunc = "mean" # default
)
```

```python
df.pivot_table(
    index = "NumberOfChildren", # row
    columns = "IsSoftwareDev",
    values = "HoursSpentLearningToCode",
    aggfunc = "count"
)
```

```python
df.pivot_table(
    index = "NumberOfChildren", # row
    columns = "IsSoftwareDev",
    values = "HoursSpentLearningToCode",
    aggfunc = "max"
)
```

```python
df.pivot_table(
    index = "NumberOfChildren",
    columns = "IsSoftwareDev",
    values = "HoursSpentLearningToCode",
    aggfunc = "min"
)
```

```python
df.pivot_table(
    index = "NumberOfChildren",
    columns = "IsSoftwareDev",
    values = "HoursSpentLearningToCode",
    aggfunc = "std"
)
```

```python
df.pivot_table(
    index = "NumberOfChildren",
    columns = "IsSoftwareDev",
    values = "HoursSpentLearningToCode",
    aggfunc = ["median", "mean", "sum", "max", "min", "std"]
)
```

## Buenos Aires Subway - Descriptive statistics


```python
df = pd.read_csv("march2024_pax_hourly.csv")
```

```python
df.head()
```

```python
# get the 0.95 quantile of the pax_TOTAL
pax_95q = df["pax_TOTAL"].quantile(0.95)

# Filter the dataset
df_95q = df[df["pax_TOTAL"] > pax_95q]

# Sort the dataset
df_95q = df_95q.sort_values(by = "pax_TOTAL", ascending = False)

df_95q.head()
```

```python
df_95q["pax_TOTAL"].describe()
```

```python
##want to know the top stations that encounter this kind of passenger traffic.

# get the counts for each station
station_counts = df_95q["station"].value_counts()

station_counts.head()
```

```python
# # Get the top 10 results

station_counts_10 = station_counts.head(10)
type(station_counts_10)
```

```python
station_counts_10 = station_counts.head(10)
station_counts_10.plot(kind = "bar")
```

##Statistics per station


```python
grouped_by_station = df_95q.groupby("station")

# mean of pax_TOTAL per station
mean_per_station = grouped_by_station["pax_TOTAL"].mean()

mean_per_station.head()
```

## Statistics per hour and subway line

want to calculate statistics both by hour and line.

```python
df_95q.pivot_table(
    index = "hour",
    columns = "line",
    values = "pax_TOTAL",
    aggfunc = ["mean", "median", "max", "min", "std"]
)
```
