```python
#df_PM25 = df_AQ[df_AQ["Name"] == 'Fine particles (PM 2.5)']

df_PM25 = df_AQ[df_AQ["Name"] == 'Fine particles (PM 2.5)']
```

```python
df_PM25.get("Geo Place Name").tolist()

geo_place_name = df_PM25.get("Geo Place Name").tolist()
data_value = df_PM25.get("Data Value").tolist()
```

```python
num_observations = len(geo_place_name)
print("Number of observations:", num_observations)
```

```python
for i in range(num_observations):
  if data_value[i] > 12:
    print(geo_place_name[i], data_value[i])
```

```python
CC_values = [] # Chelsea - Clinton
LI_values = [] # Long Island City - Astoria
UWS_values = [] # Upper West Side


for i in range(num_observations):
  if geo_place_name[i] == "Chelsea - Clinton":
    CC_values.append(data_value[i])
  elif geo_place_name[i] == "Long Island City - Astoria":
    LI_values.append(data_value[i])
  elif geo_place_name[i] == "Upper West Side":
    UWS_values.append(data_value[i])
```

```python
print(CC_values)
```

```python
print("There are ", len(CC_values), " observations for Chelsea - Clinton")
print("There are ", len(LI_values), " observations for Long Island City - Astoria")
print("There are ", len(UWS_values), " observations for Upper West Side")
```

```python
average_CC = sum(CC_values) / len(CC_values)
average_LI = sum(LI_values) / len(LI_values)
average_UWS = sum(UWS_values) / len(UWS_values)

print(average_CC, average_LI, average_UWS)
```
